```python
# identity class creator

# Creates a class that contains a new filtered dataframe that for a particular identify marker 
# and comparison type; the class also contains statistical information and a function that will print the stats

class identity:
    def __init__(self, data, label, identity_marker, comparison_type, desired_value):
        import pandas as pd
        
        df = pd.read_csv(data)
        
        
        if type(identity_marker) is list:
    
            new_df = df
            i = 0
            
            for comp in comparison_type:
                if comp == 'equal':
                    new_df = new_df[new_df[identity_marker[i]] == desired_value[i]]
                
                elif comp == 'greater than':
                    new_df = new_df[new_df[identity_marker[i]] > desired_value[i]]
                
                elif comp == 'lesser than':
                    new_df = new_df[new_df[identity_marker[i]] < desired_value[i]]
               
                elif comp == 'within range':
                    new_df = new_df[(new_df[identity_marker[i]] > desired_value[i][0]) & 
                                    (new_df[identity_marker[i]] < desired_value[i][1])]
                elif comp == 'outside of range':
                    new_df = new_df[(new_df[identity_marker[i]] < desired_value[i][0]) | 
                                    (new_df[identity_marker[i]] > desired_value[i][1])]
                
                elif comp == 'contains':
                    new_df = new_df[new_df[identity_marker[i]].str.contains(desired_value[i])]
                
                else:
                    raise Exception("invalid comparison type, try equal, greater than, lesser than, within range, outside of range, contains")
                i+=1
                
        else:
            
            if comparison_type == 'equal':
                new_df = df[df[identity_marker] == desired_value]

            elif comparison_type == 'greater than':
                new_df = df[df[identity_marker] > desired_value]

            elif comparison_type == 'lesser than':
                new_df = df[df[identity_marker] < desired_value]

            elif comparison_type == "within range":
                new_df = df[(df[identity_marker] > desired_value[0]) & (df[identity_marker] < desired_value[1])]
            
            elif comparison_type == 'outside of range':
                new_df = df[(df[identity_marker] < desired_value[0]) | (df[identity_marker] > desired_value[1])]
            
            elif comparison_type == 'contains':
                new_df = df[df[identity_marker].str.contains(desired_value)]
            
            else:
                raise Exception("invalid comparison type, try equal, grather than, lesser than, within range, outside of range, contains")

        # store inputs, new df, and statistical values

        self.id_marker = identity_marker
        self.label = label
        self.comp = comparison_type
        self.des_val = desired_value
        self.new_df = new_df.reset_index(drop=True)
        self.count = len(new_df)
        self.avg = round(self.new_df[label].mean(), 2)
        self.std_dev = round(self.new_df[label].std(),2)
        self.median = round(self.new_df[label].median(),2)
    
    
    def stat_readout(self): 
        
        if type(self.id_marker) is list:
            print("The average {} for {} with comparison types: {},{} is {} \u00B1 {}, while the median {} is {} ".format(
            self.label, self.id_marker, self.comp,self.des_val, self.avg, self.std_dev, self.label, self.median))
        else:
            if self.comp == 'within range':
                print("The average {} for {}s within a range of {} is {} \u00B1 {}, \nwhile the median {} is {}".format(
                    self.label, self.id_marker, self.des_val, self.avg, self.std_dev, self.label, self.median))
            elif self.comp == 'greater than' or self.comp == 'lesser than':
                print("The average {} for {}s {} {} is {} \u00B1 {}, \nwhile the median {} is {}".format(
                    self.label, self.id_marker, self.comp, self.des_val, self.avg, self.std_dev, self.label, self.median))
            else: 
                print("The average {} for {} {} {} is {} \u00B1 {}, \nwhile the median {} is {}".format(
                    self.label, self.id_marker, self.comp, self.des_val, self.avg, self.std_dev, self.label, self.median))
```
